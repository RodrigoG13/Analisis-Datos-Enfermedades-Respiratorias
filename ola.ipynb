```python
import pandas as pd
import folium

# Cargar los datos desde un archivo CSV
df = pd.read_csv("unique_points.csv")  # Asegúrate de reemplazar 'ruta_del_archivo.csv' con la ruta real a tu archivo

# Crear un mapa centrado en el promedio de las coordenadas
m = folium.Map(location=[df['LAT_DECIMAL'].mean(), df['LON_DECIMAL'].mean()], zoom_start=10)

# Dibujar los marcadores para cada punto
for idx, row in df.iterrows():
    # Asignar un color por cluster
    colors = ['red', 'green', 'orange', 'purple', 'black']
    color = colors[int(row['Cluster'])] if row['Cluster'] < len(colors) else 'blue'  # Usar azul como color por defecto si el índice está fuera de rango

    folium.CircleMarker(
        location=[row['LAT_DECIMAL'], row['LON_DECIMAL']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color
    ).add_to(m)

# Guardar el mapa en un archivo HTML
m
```
