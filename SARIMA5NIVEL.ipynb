### Cluster 0

```python
import pandas as pd
from scipy.stats import boxcox
from statsmodels.tsa.statespace.sarimax import SARIMAX
from scipy.special import inv_boxcox
import plotly.graph_objects as go
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error, mean_absolute_error
import numpy as np

def cargar_datos_cluster(ruta_csv, cluster_id):
    df = pd.read_csv(ruta_csv)
    df.set_index('Cluster', inplace=True)
    df = df.transpose()
    cluster_data = df[cluster_id].reset_index()
    cluster_data.columns = ['Fecha', 'Contagios']
    cluster_data['Fecha'] = pd.to_datetime(cluster_data['Fecha'], format='%d-%m-%Y')
    cluster_data.set_index('Fecha', inplace=True)
    return cluster_data

def preparar_datos(df, n_preds=10):
    df['Contagios'] = df['Contagios'].clip(lower=1)
    df['Contagios_transformed'], lambda_ = boxcox(df['Contagios'])
    train_data = df['Contagios_transformed'].iloc[:-n_preds]
    test_data = df.iloc[-n_preds:]
    return train_data, test_data, lambda_

def ajustar_modelo_sarima(train_data, order, seasonal_order):
    model = SARIMAX(train_data,
                    order=order,
                    seasonal_order=seasonal_order,
                    enforce_stationarity=False,
                    enforce_invertibility=False)
    results = model.fit()
    
    print(results.summary())
    results.plot_diagnostics(figsize=(18, 8))
    plt.show()
    
    return results

def realizar_predicciones(results, steps, test_data, lambda_):
    pred_uc = results.get_forecast(steps=steps)
    pred_mean_transformed = pred_uc.predicted_mean
    pred_mean = inv_boxcox(pred_mean_transformed, lambda_)
    pred_ci_transformed = pred_uc.conf_int()
    pred_ci = pd.DataFrame()
    pred_ci['lower'] = inv_boxcox(pred_ci_transformed.iloc[:, 0], lambda_)
    pred_ci['upper'] = inv_boxcox(pred_ci_transformed.iloc[:, 1], lambda_)
    pred_mean.index = test_data.index
    return pred_mean, pred_ci

def evaluar_metricas(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    return mse, rmse, mae, r2, mape

def visualizar_resultados(df, pred_mean, pred_ci, test_data, train_data_transformed, results, lambda_, cluster_id):
    train_pred_transformed = results.fittedvalues
    train_pred = inv_boxcox(train_pred_transformed, lambda_)
    
    train_mse, train_rmse, train_mae, train_r2, train_mape = evaluar_metricas(df['Contagios'].iloc[:-len(test_data)], train_pred)
    test_mse, test_rmse, test_mae, test_r2, test_mape = evaluar_metricas(test_data['Contagios'], pred_mean)
    
    print(f"Train MSE: {train_mse}, Train RMSE: {train_rmse}, Train MAE: {train_mae}, Train R: {train_r2}, Train MAPE: {train_mape}")
    print(f"Test MSE: {test_mse}, Test RMSE: {test_rmse}, Test MAE: {test_mae}, Test R: {test_r2}, Test MAPE: {test_mape}")
    
    error_absoluto = abs(test_data['Contagios'] - pred_mean)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df.index, y=df['Contagios'], mode='lines+markers', name='Observado'))
    fig.add_trace(go.Scatter(x=pred_mean.index, y=pred_mean, mode='lines+markers', name='Pronstico'))
    fig.add_trace(go.Scatter(x=pred_ci.index, y=pred_ci.iloc[:, 0], fill=None, mode='lines', line=dict(color='lightgrey'), showlegend=False))
    fig.add_trace(go.Scatter(x=pred_ci.index, y=pred_ci.iloc[:, 1], fill='tonexty', mode='lines', line=dict(color='lightgrey'), showlegend=False))
    fig.add_trace(go.Scatter(x=test_data.index, y=test_data['Contagios'], mode='lines+markers', name='Real', line=dict(dash='dash')))
    fig.add_trace(go.Bar(x=test_data.index, y=error_absoluto, name='Error Absoluto'))
    fig.update_layout(title=f'Grfica de prediccin - Cluster {cluster_id}',
                    xaxis_title='Fecha',
                    yaxis_title='Contagios',
                    legend_title='Leyenda')

    fig.show()

def sarima_search(train_data, p_values, d_values, q_values, P_values, D_values, Q_values, s_values):
    best_score_rmse, best_score_mae, best_cfg_rmse, best_cfg_mae = float("inf"), float("inf"), None, None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                for P in P_values:
                    for D in D_values:
                        for Q in Q_values:
                            for s in s_values:
                                order = (p, d, q)
                                seasonal_order = (P, D, Q, s)
                                try:
                                    model = SARIMAX(train_data, order=order, seasonal_order=seasonal_order,
                                                    enforce_stationarity=False, enforce_invertibility=False)
                                    results = model.fit(disp=False)
                                    yhat = results.fittedvalues
                                    rmse = np.sqrt(mean_squared_error(train_data, yhat))
                                    mae = mean_absolute_error(train_data, yhat)
                                    if rmse < best_score_rmse:
                                        best_score_rmse, best_cfg_rmse = rmse, (order, seasonal_order)
                                    if mae < best_score_mae:
                                        best_score_mae, best_cfg_mae = mae, (order, seasonal_order)
                                    print(f"SARIMA{order}x{seasonal_order} - RMSE:{rmse}, MAE:{mae}")
                                except Exception as e:
                                    continue
    print(f"Best SARIMA by RMSE: {best_cfg_rmse} - RMSE:{best_score_rmse}")
    print(f"Best SARIMA by MAE: {best_cfg_mae} - MAE:{best_score_mae}")
    return best_cfg_rmse, best_cfg_mae

# Parmetros de bsqueda
p_values = [1, 2, 3]
d_values = [1, 2 ,3]
q_values = [1, 2, 3, 4, 5]
P_values = [1, 2]
D_values = [1]
Q_values = [1, 2]
s_values = [7, 30]

ruta_csv = 'clusterSemanal2123.csv'
cluster_id = 0
n_preds = 6

df = cargar_datos_cluster(ruta_csv, cluster_id)
train_data, test_data, lambda_ = preparar_datos(df, n_preds)

best_order_rmse, best_order_mae = sarima_search(train_data, p_values, d_values, q_values, P_values, D_values, Q_values, s_values)

print("Using RMSE minimized parameters:")
results_rmse = ajustar_modelo_sarima(train_data, best_order_rmse[0], best_order_rmse[1])
pred_mean_rmse, pred_ci_rmse = realizar_predicciones(results_rmse, n_preds, test_data, lambda_)
visualizar_resultados(df, pred_mean_rmse, pred_ci_rmse, test_data, train_data, results_rmse, lambda_, cluster_id)

print("Using MAE minimized parameters:")
results_mae = ajustar_modelo_sarima(train_data, best_order_mae[0], best_order_mae[1])
pred_mean_mae, pred_ci_mae = realizar_predicciones(results_mae, n_preds, test_data, lambda_)
visualizar_resultados(df, pred_mean_mae, pred_ci_mae, test_data, train_data, results_mae, lambda_, cluster_id)
```

```python
def preparar_datos_actuales(df):
    df['Contagios'] = df['Contagios'].clip(lower=1)
    df['Contagios_transformed'], lambda_ = boxcox(df['Contagios'])
    return df['Contagios_transformed'], lambda_


def realizar_predicciones_futuras(modelo, steps, lambda_):
    pred_uc = modelo.get_forecast(steps=steps)
    pred_mean_transformed = pred_uc.predicted_mean
    pred_mean = inv_boxcox(pred_mean_transformed, lambda_)
    pred_ci_transformed = pred_uc.conf_int()
    pred_ci = pd.DataFrame()
    pred_ci['lower'] = inv_boxcox(pred_ci_transformed.iloc[:, 0], lambda_)
    pred_ci['upper'] = inv_boxcox(pred_ci_transformed.iloc[:, 1], lambda_)
    last_date = modelo.data.dates[-1]
    future_dates = pd.date_range(start=last_date + pd.DateOffset(days=1), periods=steps, freq='W')
    pred_mean.index = future_dates
    pred_ci.index = future_dates
    return pred_mean, pred_ci



def visualizar_predicciones_futuras(df, pred_mean, pred_ci, cluster_id):
    fig = go.Figure()

    fig.add_trace(go.Scatter(x=df.index, y=df['Contagios'], mode='lines+markers', name='Observado'))

    fig.add_trace(go.Scatter(x=pred_mean.index, y=pred_mean, mode='lines+markers', name='Pronstico'))

    fig.add_trace(go.Scatter(x=pred_ci.index, y=pred_ci['lower'], fill=None, mode='lines', line=dict(color='lightgrey'), showlegend=False))

    fig.add_trace(go.Scatter(x=pred_ci.index, y=pred_ci['upper'], fill='tonexty', mode='lines', line=dict(color='lightgrey'), showlegend=False))

    fig.update_layout(title=f'Prediccin de Contagios - Cluster {cluster_id}',
                    xaxis_title='Fecha',
                    yaxis_title='Contagios',
                    legend_title='Leyenda')

    fig.show()


datos = cargar_datos_cluster(ruta_csv, cluster_id)
datos_transformados, lambda_ = preparar_datos_actuales(datos)
modelo_ajustado = ajustar_modelo_sarima(datos_transformados, order, seasonal_order)
predicciones, intervalos_confianza = realizar_predicciones_futuras(modelo_ajustado, n_preds, lambda_)
visualizar_predicciones_futuras(datos, predicciones, intervalos_confianza, cluster_id)
```
